# Loading Data Set

```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data for housing
data = pd.read_csv("housing.csv")
```

```python
data
```

# Data Exploration

```python
data.info()
```

```python
#from the above information we see that there are some null values in the total_bedrooms so we will drop those values
#data.dropna() -> only drops the data doesnt save to data
data.dropna(inplace=True) #-> saves the data to data variable
```

```python
data.info()
#visualising the new values that has been saved
```

```python
#importing train_test_split to split into training and testing data
from sklearn.model_selection import train_test_split
X = data.drop(['median_house_value'],axis=1)
#x is consisting of everything except target value hence we tell axis =1
y = data['median_house_value']
#y consists of only median_house_value
```

```python
X
```

```python
y
```

```python
#splitting x and y into training and testing sets, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
#Joining the x and y training data for analysis
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,10))
```

# Data Preprocessing

```python
#data above is skewed so we use log to adjust
train_data['total_rooms'] = np.log(train_data['total_rooms']) +1 #to avoid 0
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']) +1 #to avoid 0
train_data['population'] = np.log(train_data['population']) +1 #to avoid 0
train_data['households'] = np.log(train_data['households']) +1 #to avoid 0
train_data['median_income'] = np.log(train_data['median_income']) +1 #to avoid 0
```

```python
train_data.hist(figsize=(15,8))
```

```python
#above is more balanced
train_data.ocean_proximity.value_counts()
```

```python
#pd.get_dummies(train_data.ocean_proximity) one hot encode
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
#from above we can infer that thhe price for inland is less than when near ocean as neagtive value
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude",y="longitude",data=train_data,hue="median_house_value",palette="coolwarm")
```

# Feature Engineering

```python
#Adding interesting features:
#Number of bedrooms per room:
train_data['bedroom ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
#How many rooms per household:
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

# Linear Regression Model

```python
#from above we see that our new values has a significant correlation to house value
#can drop other features as well the ones close to 0
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
#Didnt need to split x,y that early
scaler = StandardScaler()
X_train, y_train = train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)
reg = LinearRegression()

reg.fit(X_train_s,y_train)
#didnt do scaling yet(did it after got 0.619 reg score), just seeing how well it fits with test data
```

```python
#We are going to be doing same thing to test data as we did with train data
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms']) +1 #to avoid 0
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']) +1 #to avoid 0
test_data['population'] = np.log(test_data['population']) +1 #to avoid 0
test_data['households'] = np.log(test_data['households']) +1 #to avoid 0
test_data['median_income'] = np.log(test_data['median_income']) +1 #to avoid 0

test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

#Number of bedrooms per room:
test_data['bedroom ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
#How many rooms per household:
test_data['household_rooms'] = test_data['total_rooms']/test_data['households']

X_test, y_test = test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
test_data
```

```python
reg.score(X_test,y_test) #error cause feature names present it is removed after scaling
```

```python
#above score is not bad but not great
#Lets try scaling going to go back and scale
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s,y_test)
```

```python
#didnt change too much after scaling
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train_s,y_train)
```

```python
forest.score(X_test_s,y_test)
```

```python
#as you can see random forest already gives greate values without scaling
#K fold cross validation is going to be done using hyperparameter
from sklearn.model_selection import GridSearchCV
forest = RandomForestRegressor()
param_grid = {
    "n_estimators": [3,10,30],
    "max_features":[2,4,6,8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error",
                           return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_s,y_test)
```
